```python
common_symptoms = [
    "высокая температура",
    "поражение респираторной системы",
    "конъюнктивит",
    "вялость",
    "проблемы с ЖКТ",
    "отказ от пищи",
    "извращенный аппетит",
    "постоянный плач",
    "пугливость",
]

num_samples = 100_000

with open("diseases.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(all_symptoms + ["болезнь"])  # заголовок

    for _ in range(num_samples):
        disease, symptoms = random.choice(list(diseases.items()))
        row = [0] * len(all_symptoms)

        # Основные симптомы болезни
        for symptom, weight in symptoms.items():
            row[all_symptoms.index(symptom)] = 1 if random.random() < weight else 0

        # Шум: только частые симптомы
        num_disease_symptoms = len(symptoms)
        noise_prob = 0.1 * (num_disease_symptoms / 8)  # чем больше симптомов, тем выше шум
        noise_symptoms = [
            s for s in common_symptoms if s not in symptoms
        ]
        for ns in noise_symptoms:
            if random.random() < noise_prob:
                row[all_symptoms.index(ns)] = 1

        row.append(disease)
        writer.writerow(row)
```
